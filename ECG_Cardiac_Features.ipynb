```python
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import wfdb
import random
import neurokit2 as nk
```

```python
dataset = pd.read_csv('patient_scp.csv')
dataset.head()
directory = 'physionet.org/files/ptb-xl/1.0.3/'
```

```python
feature_data = open('features.csv', 'w', newline='')
feature_writer = csv.writer(feature_data)

feature_writer.writerow(['ecg_id','Patient_ID','Label','Heart Rate', 'HRV_MeanNN', 'HRV_SDNN', 'HRV_RMSSD','HRV_pNN50'])
```

```python
# https://neuropsychology.github.io/NeuroKit/examples/ecg_hrv/ecg_hrv.html
for index, row in dataset.iterrows():
    print(f"{index}/{len(dataset)}")
    patient_id = row['patient_id']
    ecg_id = row['ecg_id']
    label = row['label']

    record = wfdb.rdrecord(directory + row['filename_hr'])
    ecg_signal = record.p_signal[:, 0]

    ecg_cleaned = nk.ecg_clean(ecg_signal, sampling_rate=500)
    ecg_peaks, ecg_info = nk.ecg_peaks(ecg_cleaned, sampling_rate=500)
    r_peaks = ecg_info["ECG_R_Peaks"]
    if r_peaks.size == 0:
        print(f"No R-peaks detected for ECG ID {ecg_id}")
        continue

    rr_intervals = nk.ecg_rate(r_peaks, sampling_rate=500)
    heart_rate = np.mean(rr_intervals)
    try:
        hrv_time = nk.hrv_time(ecg_peaks, sampling_rate=500)
    except IndexError:
        print(f"HRV computation failed for ECG ID {ecg_id}")
        continue

    feature_writer.writerow([
        ecg_id,
        patient_id,
        label,
        heart_rate,
        hrv_time['HRV_MeanNN'].iloc[0],
        hrv_time['HRV_SDNN'].iloc[0],
        hrv_time['HRV_RMSSD'].iloc[0],
        hrv_time['HRV_pNN50'].iloc[0]
    ])


```

```python
hrv_dataset = pd.read_csv('features.csv')

hrv_dataset['QRS_duration'] = np.nan
hrv_dataset['PR_duration'] = np.nan
for index, row in hrv_dataset.iterrows():
    print(f"{index}/{len(hrv_dataset)}")
    ecg_id = row['ecg_id']
    record = wfdb.rdrecord(directory + dataset[dataset['ecg_id'] == ecg_id]['filename_hr'].values[0])
    ecg_signal = record.p_signal[:, 0]
    ecg_cleaned = nk.ecg_clean(ecg_signal, sampling_rate=500)
    try:
        signals, waves = nk.ecg_delineate(ecg_cleaned, rpeaks=None, sampling_rate=500, method="dwt")
    except:
        hrv_dataset.loc[index, 'PR_duration'] = np.nan
        hrv_dataset.loc[index, 'QRS_duration'] = np.nan

    qrs_onsets = waves["ECG_R_Onsets"]
    qrs_offsets = waves["ECG_R_Offsets"]

    qrs_onsets = [onset for onset in qrs_onsets if not pd.isna(onset)]
    qrs_offsets = [offset for offset in qrs_offsets if not pd.isna(offset)]

    valid_pairs = [
        (onset, offset) for onset, offset in zip(qrs_onsets, qrs_offsets)
        if not pd.isna(onset) and not pd.isna(offset) and onset < offset
    ]

    if valid_pairs:
        qrs_durations = [(offset - onset) / 500 * 1000 for onset, offset in valid_pairs]

        plausible_durations = [duration for duration in qrs_durations if 50 <= duration <= 200]

        if plausible_durations:
            mean_qrs_duration = np.mean(plausible_durations)
            print(f"Mean QRS Duration: {mean_qrs_duration:.2f} ms")
        else:
            print("No plausible QRS durations detected.")
            mean_qrs_duration = np.nan
    else:
        print("No valid QRS complexes detected.")
        mean_qrs_duration = np.nan

    hrv_dataset.loc[index, 'QRS_duration'] = mean_qrs_duration

    if "ECG_P_Onsets" in waves and "ECG_R_Onsets" in waves:
        p_onsets = waves["ECG_P_Onsets"]
        r_onsets = waves["ECG_R_Onsets"]

        p_onsets = [onset for onset in p_onsets if not pd.isna(onset)]
        r_onsets = [onset for onset in r_onsets if not pd.isna(onset)]

        valid_pr_pairs = [
            (p_onset, r_onset) for p_onset, r_onset in zip(p_onsets, r_onsets)
            if p_onset < r_onset
        ]

        if valid_pr_pairs:
            pr_durations = [(r_onset - p_onset) / 500 * 1000 for p_onset, r_onset in valid_pr_pairs]  # Convert to ms
            plausible_pr_durations = [duration for duration in pr_durations if 100 <= duration <= 200]

            if plausible_pr_durations:
                mean_pr_duration = np.mean(plausible_pr_durations)
                print(f"Mean PR Duration: {mean_pr_duration:.2f} ms")
            else:
                print("No plausible PR intervals detected.")
                mean_pr_duration = np.nan
        else:
            print("No valid PR intervals detected.")
            mean_pr_duration = np.nan
    else:
        print("Missing delineation keys for PR interval calculation.")
        mean_pr_duration = np.nan 

    hrv_dataset.loc[index, 'PR_duration'] = mean_pr_duration

hrv_dataset.to_csv('features.csv', index=False)
```

```python
hrv_dataset = pd.read_csv('features.csv')
hrv_dataset['RR Interval Mean'] = np.nan
hrv_dataset['RR Interval SD'] = np.nan
hrv_dataset['RR Interval RMSSD'] = np.nan
hrv_dataset['RR Interval Min'] = np.nan
hrv_dataset['RR Interval Max'] = np.nan
for index, row in hrv_dataset.iterrows():
    print(f"{index}/{len(hrv_dataset)}")
    ecg_id = row['ecg_id']
    record = wfdb.rdrecord(directory + dataset[dataset['ecg_id'] == ecg_id]['filename_hr'].values[0])
    ecg_signal = record.p_signal[:, 0]
    ecg_cleaned = nk.ecg_clean(ecg_signal, sampling_rate=500)
    _, rpeaks = nk.ecg_peaks(ecg_cleaned, sampling_rate=500)
    rpeaks_indices = rpeaks['ECG_R_Peaks']

    rr_intervals_ms = np.diff(rpeaks_indices) / 500 * 1000  # Sampling rate is 500 Hz
        
    rr_mean = np.mean(rr_intervals_ms)
    rr_sd = np.std(rr_intervals_ms)
    rr_min = np.min(rr_intervals_ms)
    rr_max = np.max(rr_intervals_ms)
    
    rr_diff = np.diff(rr_intervals_ms)
    rmssd = np.sqrt(np.mean(rr_diff**2))
    
    hrv_dataset.loc[index, 'RR Interval Mean'] = rr_mean
    hrv_dataset.loc[index, 'RR Interval SD'] = rr_sd
    hrv_dataset.loc[index, 'RR Interval RMSSD'] = rmssd
    hrv_dataset.loc[index, 'RR Interval Min'] = rr_min
    hrv_dataset.loc[index, 'RR Interval Max'] = rr_max

hrv_dataset.to_csv('features.csv', index=False)

```

```python
import pandas as pd
import numpy as np
import wfdb
import neurokit2 as nk

# Load your dataset
hrv_dataset = pd.read_csv('features.csv')

# Initialize new columns for QT interval and ST segment features
hrv_dataset['QT Interval Mean'] = np.nan
hrv_dataset['QT Interval SD'] = np.nan
hrv_dataset['ST Segment Amplitude Mean'] = np.nan
hrv_dataset['ST Segment Amplitude SD'] = np.nan
hrv_dataset['ST Segment Duration Mean'] = np.nan
hrv_dataset['ST Segment Duration SD'] = np.nan


for index, row in hrv_dataset.iterrows():
    print(f"Processing record {index+1}/{len(hrv_dataset)}")
    ecg_id = row['ecg_id']
    
    # Retrieve the filename for the current ecg_id
    filename_row = dataset[dataset['ecg_id'] == ecg_id]
    if filename_row.empty:
        print(f"No matching filename for ecg_id {ecg_id}")
        continue
    filename = filename_row['filename_hr'].values[0]
    
    # Load the ECG record
    record = wfdb.rdrecord(directory + filename)
    ecg_signal = record.p_signal[:, 0]  # Assuming ECG is in the first channel
    
    # Clean the ECG signal
    ecg_cleaned = nk.ecg_clean(ecg_signal, sampling_rate=500)
    
    # Detect R-peaks
    signals, info = nk.ecg_peaks(ecg_cleaned, sampling_rate=500)
    rpeaks_indices = info['ECG_R_Peaks']
    
    # Ensure there are enough R-peaks to proceed
    if len(rpeaks_indices) > 1:
        # Delineate ECG waves to find wave boundaries
        try:
            # Delineate the ECG signal to find PQRST complexes
            signals, waves = nk.ecg_delineate(ecg_cleaned, rpeaks_indices, sampling_rate=500, method="dwt")
            
            # Extract the onsets and offsets using available keys
            r_onsets = waves['ECG_R_Onsets']
            r_offsets = waves['ECG_R_Offsets']
            t_onsets = waves['ECG_T_Onsets']
            t_offsets = waves['ECG_T_Offsets']
            
            # Convert lists to numpy arrays for element-wise operations
            r_onsets = np.array(r_onsets)
            r_offsets = np.array(r_offsets)
            t_onsets = np.array(t_onsets)
            t_offsets = np.array(t_offsets)
            
            # Ensure all arrays are of the same length
            min_length = min(len(r_onsets), len(r_offsets), len(t_onsets), len(t_offsets))
            r_onsets = r_onsets[:min_length]
            r_offsets = r_offsets[:min_length]
            t_onsets = t_onsets[:min_length]
            t_offsets = t_offsets[:min_length]
            
            # Initialize lists to store intervals and amplitudes
            qt_intervals = []
            st_durations = []
            st_amplitudes = []
            
            # Get signal length for bounds checking
            signal_length = len(ecg_cleaned)
            
            # Iterate over each beat to calculate intervals and amplitudes
            for i in range(min_length):
                r_onset = r_onsets[i]
                t_offset = t_offsets[i]
                r_offset = r_offsets[i]
                t_onset = t_onsets[i]
                
                # Check for valid QT interval
                if not np.isnan(r_onset) and not np.isnan(t_offset):
                    # Convert to integers
                    r_onset_int = int(r_onset)
                    t_offset_int = int(t_offset)
                    if 0 <= r_onset_int < t_offset_int <= signal_length:
                        # Calculate QT interval
                        qt_interval = (t_offset_int - r_onset_int) / 500 * 1000  # Convert to milliseconds
                        qt_intervals.append(qt_interval)
                    else:
                        qt_intervals.append(np.nan)
                else:
                    qt_intervals.append(np.nan)
                
                # Check for valid ST segment
                if not np.isnan(r_offset) and not np.isnan(t_onset):
                    # Convert to integers
                    r_offset_int = int(r_offset)
                    t_onset_int = int(t_onset)
                    if 0 <= r_offset_int < t_onset_int <= signal_length:
                        # Calculate ST duration
                        st_duration = (t_onset_int - r_offset_int) / 500 * 1000
                        st_durations.append(st_duration)
                        
                        # Extract ST segment
                        st_segment = ecg_cleaned[r_offset_int:t_onset_int]
                        if len(st_segment) > 0:
                            # Calculate mean amplitude of the ST segment
                            st_amplitude = np.mean(st_segment)
                            st_amplitudes.append(st_amplitude)
                        else:
                            st_amplitudes.append(np.nan)
                    else:
                        st_durations.append(np.nan)
                        st_amplitudes.append(np.nan)
                else:
                    st_durations.append(np.nan)
                    st_amplitudes.append(np.nan)
            
            # Compute statistical features of QT intervals
            valid_qt_intervals = [qt for qt in qt_intervals if not np.isnan(qt)]
            if len(valid_qt_intervals) > 0:
                qt_mean = np.mean(valid_qt_intervals)
                qt_sd = np.std(valid_qt_intervals)
            else:
                qt_mean = np.nan
                qt_sd = np.nan
            
            # Compute statistical features of ST segment durations
            valid_st_durations = [dur for dur in st_durations if not np.isnan(dur)]
            if len(valid_st_durations) > 0:
                st_dur_mean = np.mean(valid_st_durations)
                st_dur_sd = np.std(valid_st_durations)
            else:
                st_dur_mean = np.nan
                st_dur_sd = np.nan
            
            # Compute statistical features of ST segment amplitudes
            valid_st_amplitudes = [amp for amp in st_amplitudes if not np.isnan(amp)]
            if len(valid_st_amplitudes) > 0:
                st_amp_mean = np.mean(valid_st_amplitudes)
                st_amp_sd = np.std(valid_st_amplitudes)
            else:
                st_amp_mean = np.nan
                st_amp_sd = np.nan
            
            # Store the features in the DataFrame using .loc
            hrv_dataset.loc[index, 'QT Interval Mean'] = qt_mean
            hrv_dataset.loc[index, 'QT Interval SD'] = qt_sd
            hrv_dataset.loc[index, 'ST Segment Amplitude Mean'] = st_amp_mean
            hrv_dataset.loc[index, 'ST Segment Amplitude SD'] = st_amp_sd
            hrv_dataset.loc[index, 'ST Segment Duration Mean'] = st_dur_mean
            hrv_dataset.loc[index, 'ST Segment Duration SD'] = st_dur_sd
            
        except Exception as e:
            print(f"Error processing ECG ID {ecg_id}: {e}")
            # Store NaN values in case of error
            hrv_dataset.loc[index, ['QT Interval Mean', 'QT Interval SD',
                                    'ST Segment Amplitude Mean', 'ST Segment Amplitude SD',
                                    'ST Segment Duration Mean', 'ST Segment Duration SD']] = np.nan
    else:
        print(f"Insufficient R-peaks detected for ecg_id {ecg_id}")
        # Store NaN if not enough R-peaks
        hrv_dataset.loc[index, ['QT Interval Mean', 'QT Interval SD',
                                'ST Segment Amplitude Mean', 'ST Segment Amplitude SD',
                                'ST Segment Duration Mean', 'ST Segment Duration SD']] = np.nan

# Save the updated dataset
hrv_dataset.to_csv('features_updated.csv', index=False)
```
